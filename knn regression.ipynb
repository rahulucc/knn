```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Rahul\Documents\Desktop\MARCUS\train.csv')
df.head()
```

```python
df.isnull().sum()
#missing values in Item_weight and Outlet_size needs to be imputed
mean = df['Product_Category_2'].mean() #imputing item_weight with mean
df['Product_Category_2'].fillna(mean, inplace =True)

mode = df['Product_Category_3'].mode() #imputing outlet size with mode
df['Product_Category_3'].fillna(mode[0], inplace =True)
```

```python
df.drop(['Product_Category_2', 'Product_Category_3'], axis=1, inplace=True)
df = pd.get_dummies(df)
```

```python
from sklearn.model_selection import train_test_split
train , test = train_test_split(df, test_size = 0.3)

x_train = train.drop('Item_Outlet_Sales', axis=1)
y_train = train['Item_Outlet_Sales']

x_test = test.drop('Item_Outlet_Sales', axis = 1)
y_test = test['Item_Outlet_Sales']
```
